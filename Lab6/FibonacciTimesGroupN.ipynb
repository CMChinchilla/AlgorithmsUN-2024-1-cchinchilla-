```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fibRecursive(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibRecursive(n-1)+fibRecursive(n-2)

def fibIterative(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

def fibIterativeLogaritmic(n):
   if n==0: return 0


   v1, v2, v3 = 1, 1, 0

   for rec in bin(n)[3:]:
        calc = (v2*v2)
        v1, v2, v3 = (v1*v1+calc) , ((v1+v3)*v2) , (calc+v3*v3)
        if rec == '1': v1, v2, v3 = (v1+v2), v1, v2
   return v2
```

```python
fibRecursiveTime=[]
fibIterativeTime=[]

fibIterativeLogaritmicTime=[]


for i in range (0,46):
    begin = time.time()
    fibRecursive(i)
    end = time.time()
    timeElapsed = end-begin
    fibRecursiveTime.append(timeElapsed)

for i in range (0,1001):
    begin = time.time()
    fibIterative(i)
    end = time.time()
    timeElapsed = end -begin
    fibIterativeTime.append(timeElapsed)


for i in range (0,1001):
    begin = time.time()
    fibIterative(i)
    end = time.time()
    timeElapsed = end - begin
    fibIterativeLogaritmicTime.append(timeElapsed)



b = np.array(range(0,46) )
c = np.array(range(0,1001))


plt.plot(b,fibRecursiveTime)
plt.title("Recursive time given n")
plt.xlabel("n size")
plt.ylabel("Seconds")
plt.show()

plt.plot(c,fibIterativeTime)
plt.title("Iterative time given n")
plt.xlabel("n size")
plt.ylabel("Seconds")
plt.show()

plt.plot(c,fibIterativeLogaritmicTime)
plt.title("Iterative logaritmic time given n")
plt.xlabel("n size")
plt.ylabel("Seconds")
plt.show()
```
